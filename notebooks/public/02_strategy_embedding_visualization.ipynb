# Strategy Embedding Visualization

Small public visualization for named heuristic strategy embeddings.

```python
import matplotlib.pyplot as plt
import torch
from pathlib import Path
import sys

root = Path.cwd().resolve().parent.parent
src = root / "src"
sys.path.insert(0, str(src))

from strategy_games.strategies.embeddings import available_heuristic_strategies, named_strategy_embedding

labels = available_heuristic_strategies()
embeddings = torch.stack([named_strategy_embedding(label, 8).vector for label in labels])
plt.figure(figsize=(5, 4))
plt.scatter(embeddings[:, 0], embeddings[:, 1])
for label, xy in zip(labels, embeddings[:, :2]):
    plt.annotate(label, (float(xy[0]), float(xy[1])))
plt.xlabel('directness')
plt.ylabel('evasiveness')
plt.title('Named strategy prototypes')
plt.show()
```
